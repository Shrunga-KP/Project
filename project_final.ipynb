Column descriptions

```python
column_descriptions = {
    'crim': 'Per capita crime rate by town',
    'zn': 'Proportion of residential land zoned for large lots',
    'indus': 'Proportion of non-retail business acres per town',
    'chas': 'Charles River dummy variable (1 if tract bounds river, 0 otherwise)',
    'nox': 'Nitrogen oxide concentration (parts per 10 million)',
    'rm': 'Average number of rooms per dwelling',
    'age': 'Proportion of owner-occupied units built before 1940',
    'dis': 'Weighted distance to employment centers',
    'rad': 'Index of accessibility to radial highways',
    'tax': 'Property tax rate',
    'ptratio': 'Pupil-teacher ratio',
    'black': 'Proportion of residents of African American descent',
    'lstat': 'Percentage of lower status population',
    'medv': 'Median value of owner-occupied homes'
}
```

Finding correlation for 50% dataset

```python
num_rows_to_select = int(len(df) * 0.5)
random.seed(42)  # Set a seed for reproducibility

max_correlation_coefficients = {}  # Initialize an empty dictionary to store max coefficients

for col in df.columns[1:-1]:  # Iterate through all columns except the last one ('medv') #medv(target variable)
    max_correlation_coefficient = float("-inf")  # Initialize with negative infinity
    for _ in range(1000):  # Run 1000 iterations
        selected_rows = random.sample(range(len(df)), num_rows_to_select) #Randomly samples rows without replacement

        col_data = df.loc[selected_rows, col].to_numpy() #Extracts the values of the current feature for the selected rows
        medv_data = df.loc[selected_rows, 'medv'].to_numpy() #Extracts the values of the target variable ('medv') for the selected rows

        correlation_coefficient, p_value = pearsonr(col_data, medv_data) #Calculates the Pearson correlation coefficient

        # Update max_correlation_coefficient if a higher value is found
        if correlation_coefficient > max_correlation_coefficient:
            max_correlation_coefficient = correlation_coefficient

    max_correlation_coefficients[column_descriptions[col]] = max_correlation_coefficient

print("Maximum Pearson Correlation Coefficients:")
for col, max_corr in max_correlation_coefficients.items():
    print(f"{col} vs. medv: {max_corr}")
```

```python
random.seed(42)  # Set a seed for reproducibility

# Initialize variables to track the maximum correlation
max_correlation = float("-inf")
best_subset_percentage = None

# Iterate through a range of percentages from 0.5 to 0.9
for percentage in range(50, 100, 10): #the range starts at 50, stops before 100, and increments by 10
    num_rows_to_select = int(len(df) * percentage / 100)
    max_corr_for_percentage = float("-inf")  # Initialize with negative infinity

    for col in df.columns[1:-1]:
        max_correlation_coefficient = float("-inf")  # Initialize with negative infinity #to find the maximum value in a series of comparisons

        for _ in range(1000):  # Run 1000 iterations
            selected_rows = random.sample(range(len(df)), num_rows_to_select)

            col_data = df.loc[selected_rows, col].to_numpy()
            medv_data = df.loc[selected_rows, 'medv'].to_numpy()

            correlation_coefficient, p_value = pearsonr(col_data, medv_data)

            if correlation_coefficient > max_correlation_coefficient:
                max_correlation_coefficient = correlation_coefficient

        if max_correlation_coefficient > max_corr_for_percentage:
            max_corr_for_percentage = max_correlation_coefficient

    if max_corr_for_percentage > max_correlation:
        max_correlation = max_corr_for_percentage
        best_subset_percentage = percentage

# Print the subset with the highest correlation
print(f"The subset with the highest correlation with 'medv' is at {best_subset_percentage}% of data.")
print(f"The maximum correlation coefficient is {max_correlation:.2f}")
```
